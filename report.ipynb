### Cel ćwiczenia:

Celem ćwiczenia jest zapoznanie się z pracą z korpusami językowymi i sposobem przetwarzania ich, w celu przygotowania tokenów.

### Wykorzystane narzędzia:

Visual Studio Code, Python z bibliotekami re, math, warnings, collections, itertools, numpy, pandas, matplotlib, datasets

```python
import re, math, warnings
warnings.filterwarnings('ignore')
from collections import Counter
from itertools import islice
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

wiki = load_dataset('wikimedia/wikipedia', '20231101.pl', split='train', streaming=True)
articles = list(islice(wiki, 200))
texts  = [a['text']  for a in articles]
titles = [a['title'] for a in articles]
print(f"Załadowano {len(texts)} artykułów.")
```

### Zadanie 1

```python
print([titles[i] for i in range(10)])

for i in range(10):
    print(f"Tytuł: {titles[i]}, liczba słów: {len(texts[i])}")
    dwa_zdania = [texts[i].split(".", 1)]
    print(f"Pierwsze 2 zdania: {dwa_zdania[0]}")
```

Korpus językowy jest zbiorem tekstów, natomiast zbiór danych do klasyfikacji może przyjmować różne formy (tekst, obrazy, liczby) i przypisane ma do nich etykiety, które model ma przewidywać. Korpus językowy wykorzystywany jest do uczenia maszynowego rozpoznawania tekstów i sposobu pisania. Zbiór danych do klasyfikacji stosowany jest do uczenia podejmowania decyzji.

### Zadanie 2

```python
full_text = " ".join(texts).lower()

clean_text = re.sub(r'[^\w\s]', '', full_text)
tokens = clean_text.split()

total_tokens = len(tokens)
unique_tokens = len(set(tokens))
ttr = unique_tokens / total_tokens if total_tokens > 0 else 0

print(f"Łączna liczba tokenów: {total_tokens}")
print(f"Liczba unikalnych tokenów (types): {unique_tokens}")
print(f"TTR (Type-Token Ratio): {ttr:.4f}")

counts = Counter(tokens)
sorted_counts = counts.most_common()

print("\n20 najczęstszych słów:", sorted_counts[:20])
print("\n20 najrzadszych słów:", sorted_counts[-20:])

frequencies = [count for word, count in sorted_counts]
ranks = range(1, len(frequencies) + 1)

plt.figure(figsize=(10, 6))
plt.loglog(ranks, frequencies, marker=".", linestyle="none")
plt.title("Prawo Zipfa (skala logarytmiczna)")
plt.xlabel("Ranga (log)")
plt.ylabel("Częstotliwość (log)")
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.show()
```

Prawo Zipfa mówi, że niewielka liczba słów występuje bardzo często, a ogromna większość bardzo rzadko. W praktyce dla budżetu tokenów w LLM oznacza to, że duża część kontekstu będzie wypełniona powtarzalnymi, częstymi tokenami (np. spójnikami), które niosą mniej informacji niż rzadkie słowa, dlatego ważne jest kondensowanie tekstu w celu niemarnowania tokenów na nieważne fragmenty.


### Zadanie 3

```python
articles_length = [len(text.split()) for text in texts]
print(articles_length)

min_val = np.min(articles_length)
max_val = np.max(articles_length)
mean_val = np.mean(articles_length)
median_val = np.median(articles_length)
std_val = np.std(articles_length)
p95_val = np.percentile(articles_length, 95)

print(f"Statystyki długości artykułów (w słowach):")
print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Średnia: {mean_val:.2f}")
print(f"Mediana: {median_val}")
print(f"Odchylenie standardowe: {std_val:.2f}")
print(f"P95 (95. percentyl): {p95_val:.2f}")

plt.figure(figsize=(10, 6))
plt.hist(articles_length, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

legth = [i for i in articles_length if i*1.3 > 512]
print(len(legth))
```
